```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [10,20,30,40]
```

```python
plt.plot(x,y)
plt.title("Simple PLot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x = [1,2,3,4]
y1 = [10,20,30,40]
y2 = [15,18,22,28]
plt.plot(x,y1,label = "Line 1")
plt.plot(x,y2,label = "Line 2")
plt.legend()
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,'r--')
plt.title("Plot 1")

plt.subplot(1,2,2)
plt.plot(x,y2,'g--')
plt.title("PLot 2")

plt.show()
```

```python
categories = ['A','B','C','D']
values = [3,7,8,5]
plt.bar(categories,values,color = "Blue")
plt.title("Bar PLot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
x = [5,7,8,7,2,17]
y = [99, 86, 87, 88, 100, 86]
plt.scatter(x,y,color="Red")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.show()
```

```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
plt.hist(data,bins = 5, color = "Purple",edgecolor = "black")
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

```python
labels = ['A','B','C','D']
sizes = [15,30,45,10]
plt.pie(sizes,labels = labels,autopct='%1.1f%%',startangle=270)
plt.title("Pie Chart")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]
z = [5, 15, 25, 35]
ax.scatter(x, y, z, color='red')
ax.set_title("3D Scatter Plot")
plt.show()
```

```python
import pandas as pd
import numpy as np

dates = pd.date_range(start="2024-01-01", periods=100)
values = np.random.randn(100).cumsum()
data = pd.DataFrame({"Date": dates, "Value": values})
plt.figure(figsize=(12, 6))
plt.plot(data["Date"], data["Value"])
plt.title("Time-Series Visualization")
plt.xlabel("Date")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
import matplotlib.animation as animation

fig, ax = plt.subplots()
x, y = [], []
line, = ax.plot([], [], 'b-')

def init():
    ax.set_xlim(0, 10)
    ax.set_ylim(-1, 1)
    return line,

def update(frame):
    x.append(frame)
    y.append(np.sin(frame))
    line.set_data(x, y)
    return line,

ani = animation.FuncAnimation(fig, update, frames=np.linspace(0, 10, 100), init_func=init, blit=True)
plt.show()
```
