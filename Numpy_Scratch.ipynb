```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print(np.linalg.inv(A))
```

```python
eigvals, eigvecs = np.linalg.eig(A)
print(eigvals)
```

```python
A = np.array([[3, 1], [1, 2]])
b = np.array([9, 8])
x = np.linalg.solve(A, b)
print(x)
```
