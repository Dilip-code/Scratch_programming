```python
import numpy as np
from scipy import linalg

# Define matrix A and vector b
A = np.array([[3, 2], [6, 4]])  # This is a singular matrix
b = np.array([5, 10])

# Check the determinant of A
determinant = linalg.det(A)
print("Determinant of A:", determinant)

if determinant == 0:
    print("Matrix A is singular. Attempting least-squares solution.")
    # Use least-squares solution for singular matrix
    x, residuals, rank, singular_values = linalg.lstsq(A, b)
    print("Least-squares solution:", x)
else:
    # Solve the linear system
    x = linalg.solve(A, b)
    print("Solution:", x)
```

```python
from scipy.interpolate import interp1d
x = np.linspace(0, 10, 10)
y = np.sin(x)

f = interp1d(x, y, kind='cubic')
x_new = np.linspace(0, 10, 50)
y_new = f(x_new)
```

```python
import matplotlib.pyplot as plt

# Plot original data points
plt.plot(x, y, 'o', label='Original Data')

# Plot interpolated data
plt.plot(x_new, y_new, '-', label='Cubic Interpolation')

plt.legend()
plt.show()
```
