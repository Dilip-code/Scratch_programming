```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
sns.scatterplot(data=df,x = 'total_bill',y = 'tip',hue = 'time')
plt.title("Scatter plot tips vs bill")
plt.show()
```

```python
sns.lineplot(data = df,x = 'size',y = 'tip',hue='day')
plt.title("Line plot of tips by size")
plt.show()
```

```python
sns.barplot(data = df, x = 'day',y = 'total_bill',hue = 'sex')
plt.title("Average total bill by day and Gender")
plt.show()
```

```python
sns.histplot(data = df , x = 'total_bill', kde=True,bins = 20)
plt.title("Histogram of total bills")
plt.show()
```

```python
sns.boxplot(data = df, x = 'day',y= 'total_bill',hue = 'time')
plt.title("box plot of total bill by day")
plt.show()
```

```python
sns.violinplot(data = df, x = 'day',y= "total_bill",hue = "time",split=True)
plt.title("violin plot total bill by" )
plt.show()
```

```python
sns.set_palette("coolwarm")
sns.barplot(data=df,x = 'day',y='tip',hue= 'sex')
plt.title("Customized color palette")
plt.show()
```

```python
plot = sns.boxplot(data=df,x='day',y='total_bill')
plot.set_title("Box plot of totalbill by day")
plot.set_xlabel("Day of the week")
plot.set_ylabel("Total Bill")
plt.show()
```

```python
sns.pairplot(df,hue='sex')
plt.show()
```

```python
import pandas as pd
```

```python
df_encoded = df.copy()
categorical_cols = ['sex', 'smoker', 'day', 'time']
df_encoded = pd.get_dummies(df_encoded, columns=categorical_cols, drop_first=True)
```

```python
correlation_matrix = df_encoded.corr()
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
g = sns.FacetGrid(df,col='time',row='sex',margin_titles=True)
g.map(sns.scatterplot,'total_bill','tip')
plt.show()
```

```python
g = sns.PairGrid(df,hue='sex')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
plt.show()
```

```python
sns.set_theme(style="whitegrid")
sns.boxplot(data = df,x='day',y = 'total_bill')
plt.show()
```

```python
plot = sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plot.set_title("Correlation Heatmap")
plt.show()
```
